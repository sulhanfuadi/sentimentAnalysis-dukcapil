```python
# importing the google_play_scraper library to access reviews and app information from the Google Play Store.
from google_play_scraper import app, reviews_all, Sort

# Retrieving all reviews from the app with the ID 'gov.dukcapil.mobile_id' on the Google PlayStore.
# The scraping process may take some time depending on the number of reviews available.
scrapreview = reviews_all(
    'gov.dukcapil.mobile_id',
    lang='id', 
    country='id', 
    sort=Sort.MOST_RELEVANT, 
    count=2000, 
)
```

```python
# Creating a DataFrame from scrapreview results
app_reviews_df = pd.DataFrame(np.array(scrapreview),columns=['review'])
app_reviews_df = app_reviews_df.join(pd.DataFrame(app_reviews_df.pop('review').tolist()))

app_reviews_df
```
